```python
import pandas as pd
import json
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import pickle

pd.set_option('display.max_rows', None)
```

Load Datasets

```python
df4 = pd.read_csv('df4.csv')
df5 = pd.read_csv('df5.csv')
#df7 = pd.read_csv('df7.csv')

df8 = pd.read_csv('df8.csv')
df9 = pd.read_csv('df9.csv')
df10 = pd.read_csv('df10.csv')
```

```python
df_complete = pd.read_csv('df_complete.csv', header=None)
df_complete.drop(columns = 0, inplace = True)
df_complete.columns = ['Date', 'Time', 'Humidity','Humidity_slope','Windspeed','Windspeed_slope','Temperature',\
                       'Temperature_slope', 'Label']
```

Combine all the dataframes with labeled time series data together for creating an XGBoost model

```python
df = pd.concat([df4, df5, df8, df9, df10], axis=0)
df.reset_index(drop = True, inplace = True)
```

Run the XGBClassifier model for multiclass classification 

```python
# Humidity, Temperature2


X = df[['Humidity','Humidity_slope','Windspeed','Temperature','Temperature_slope']]
y = df[['Label']]

test_size = 0.30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_size, random_state=4)
bst = XGBClassifier(n_estimators=75,objective='reg:squarederror', max_depth = 3)
bst.fit(X_train, y_train)

y_hat_bst = bst.predict(X_test)

x = y_test; y = y_hat_bst
print("R squared:",str(round(100*r2_score(x,y),3))+'%');
```

Save model 

```python
with open('weather_model.pkl', 'wb') as file:
    pickle.dump(bst, file)
```
